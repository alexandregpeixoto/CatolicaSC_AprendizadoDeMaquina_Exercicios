```python
from auto_ts import auto_timeseries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carrega a base
dataset = pd.read_csv("/content/daily-min-temperatures.csv")
dataset.head()
```

```python
# Especifica as colunas Date e Target
date_column = "Date"
target_column = "Temp"

#Converte "Date" para formato datetime
dataset[date_column] = pd.to_datetime(dataset[date_column])
```

```python
plt.figure(figsize=(12,6))
plt.plot(dataset[date_column].values, dataset[target_column].values, marker="o", linestyle="-", color="b")
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Time series ")
plt.legend()
plt.grid(True)
plt.show()
```

```python
automl_model = auto_timeseries(
    score_type="rmse",
    time_column="datetime_column",
    target="target_column",
    forecast_period=10,
    model_type="best",
    verbose=2,
)
```

```python
automl_model.fit(traindata=dataset, ts_column=date_column, target=target_column)
```
